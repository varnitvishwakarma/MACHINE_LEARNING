handling outlier 

 5 number summary 

1. minimum value 
2. Q1 - 25 percentile 
3. median 
4. Q3 - 75 percentile 
5. maximum value 

```python
import numpy as np

list_marks=[45,85,95,25,26,85,95,55,85,57,75,85,65,45,85,99,47,64,1000,1111]
```

```python
## lower fence <--> higher fence

Q1=np.percentile(list_marks,[25])
Q3=np.percentile(list_marks,[75])
```

```python
minimum,Q1,Q2,Q3,maximum=np.quantile(list_marks,[0,0.25,0.50,0.75,1.0])
```

```python
IQR=Q3-Q1
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python
lower_fence,higher_fence
```

```python
outliers=[]
for i in list_marks:
    if i <=lower_fence or i>=higher_fence:
        
        outliers.append(i)
    else:
        print("the element is not a outlier")
```
