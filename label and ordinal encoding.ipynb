LABEL ENCODING

```python
import pandas as pd
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.DataFrame({"color":("red","blue","black","blue","red","black")})
```

```python
#create an instance of label encoder
encoder=LabelEncoder()
```

```python
encoder.fit_transform(df["color"])
```

ORDINAL ENCODING

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({"size":("small","medium","large","medium","small","large")})
```

```python
#create an instance of ordinal encoder
encoder=OrdinalEncoder(categories=[["small","medium","large"]])

encoder.fit_transform(df[["size"]])
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")

df.head()
df["class"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder=OrdinalEncoder(categories=[[ 'Third', 'Second','First']])
```

```python
num_class=encoder.fit_transform(df[["class"]])
```

```python
df_num_class=pd.DataFrame(num_class,columns=["num_class"])
```

```python
new_df=df.drop("class",axis=1)
```

```python
pd.concat([new_df,df_num_class],axis=1)
```
