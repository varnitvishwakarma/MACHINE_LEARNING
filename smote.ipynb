```python
from sklearn.datasets import make_classification
import pandas as pd 
```

```python
#x=independent feature
#y=dependent feature
x,y=make_classification(n_samples=1000,n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.90],random_state=42)
```

```python
x,y
```

```python
df1=pd.DataFrame(x,columns=["feature1","feature2"])
```

```python
df2=pd.DataFrame(y,columns=["target"])
```

```python
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df["target"].value_counts()
```

```python
import seaborn as sns
```

```python
sns.displot(final_df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(final_df["feature1"],final_df["feature2"],c=final_df["target"])
```

```python
from imblearn.over_sampling import SMOTE 
```

```python
oversample=SMOTE()
x,y=oversample.fit_resample(final_df[["feature1","feature2"]],final_df["target"])
```

```python
x.shape
```

```python
y.value_counts()
```

```python
df=pd.DataFrame(x)
df3=pd.DataFrame(y)
final_data=pd.concat([df,df3],axis=1)
```

```python
plt.scatter(final_data["feature1"],final_data["feature2"],c=final_data["target"])
```
