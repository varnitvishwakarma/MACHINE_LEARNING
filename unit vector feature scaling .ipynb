```python
import seaborn as sns
```

```python
df=sns.load_dataset("iris")
```

```python
df.head()
```

```python
from sklearn.preprocessing import normalize
```

```python
unit_vector=normalize(df[["sepal_length","sepal_width","petal_length","petal_width"]])
```

```python
import pandas as pd
pd.DataFrame(unit_vector,columns=["sepal_length","sepal_width","petal_length","petal_width"])
```
